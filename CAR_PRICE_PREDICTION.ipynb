```python
df.dtypes
```

```python
df1=pd.get_dummies(df,columns=['Car_name','fueltype','aspiration','carbody','drivewheel','enginelocation','enginetype','fuelsystem'],dtype=int)
df1
```

```python
x=df1.drop(['price'],axis=1)
```

```python
y=df1['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score,mean_absolute_percentage_error

model1=KNeighborsRegressor()
model2=LinearRegression()
model3=RandomForestRegressor()
model4=DecisionTreeRegressor()

lst=[model1,model2,model3,model4]

for i in lst:
  print('model started',i)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print('r2 score :',r2_score(y_test,y_pred))
  print('% error :',mean_absolute_percentage_error(y_test,y_pred))
  print('model ended')
  print()
```
